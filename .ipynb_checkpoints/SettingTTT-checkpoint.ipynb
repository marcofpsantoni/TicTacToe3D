```python
#Inizializzazione Tic Tac Toe 4x4x4
dim = int(4)

def inizializza():    
    d3 = [[[0 for col in range(dim)] for row in range(dim)] for x in range(dim)]
    for x in range(len(d3)):
        print("")
        for col in range(len(d3)):
            print(d3[x][col])        
    return d3
#inizializza() #Inizialization 4x4x4 with a decent graphic
```

```python
#Utile check

def checker(check):
    if (check[0]==1 or check[0]==2) and check.count(check[0])==dim:
        return True
    return False
```
